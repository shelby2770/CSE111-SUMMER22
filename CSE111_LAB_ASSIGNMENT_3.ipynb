```python
#TASK 1
class Calculator:
    def __init__(self):
        pass
    
    def add(self, num_1, num_2):
        return f"Result: {num_1 + num_2}"
    
    def subtract(self, num_1, num_2):
        return f"Result: {num_1 - num_2}"
    
    def multiply(self, num_1, num_2):
        return f"Result: {num_1 * num_2}"
    
    def divide(self, num_1, num_2):
        if num_2 == 0:return "Undefined"
        return f"Result: {num_1 / num_2}"


obj = Calculator()

num_1 = int(input())
operator = input()
num_2 = int(input())

print("Let’s Calculate!")
print("Value 1:", num_1)
print("Operator:", operator)
print("Value 2:", num_2)

if operator == "+":
    print(obj.add(num_1, num_2))
elif operator == "-":
    print(obj.subtract(num_1, num_2))
elif operator == "*":
    print(obj.multiply(num_1, num_2))
elif operator == "/":
    print(obj.divide(num_1, num_2))
else:
    print("invalid input")
```

```python
#TASK 5
class Shape:
    def __init__(self, name, side_1, side_2):
        self.name= name
        self.side_1= side_1
        self.side_2= side_2

    def area(self):
        if self.name== "Triangle":
            print(f"Area: {0.5*self.side_1*self.side_2}")   
        elif self.name== "Square" or self.name=="Rectangle":
            print(f"Area: {self.side_1*self.side_2}")     
        elif self.name== "Rhombus":
             print(f"Area: {0.5*self.side_1*self.side_2}")    
        else:
            print("Area: Shape unknown")     

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()                
```

```python
#TASK 6
class Calculator:
    def __init__(self):

        print("Calculator is ready!")

    def calculate(self,num_1, num_2, operator):    
        self.num_1= num_1
        self.num_2= num_2
        self.operator= operator

        if self.operator== "+":
            self.val= self.num_1+self.num_2
        elif self.operator== "-": 
            self.val= self.num_1-self.num_2   
        elif self.operator== "*":  
            self.val= self.num_1*self.num_2  
        elif self.operator== "/":
            self.val= self.num_1/self.num_2

        return self.val   

    def showCalculation(self):
        print(f"{self.num_1} {self.operator} {self.num_2} = {self.val}")

                            

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#TASK 7
class Student:
    def __init__(self, name, id, dept, cg_list):
        self.name= name
        self.id= id
        self.dept= dept
        self.cg_list= cg_list

    def calculate_CGPA(self):
        sum= 0
        for i in self.cg_list:
            sum+= i*3
        self.cgpa= sum/(len(self.cg_list)*3)

    def print_details(self):
        if self.cgpa>3.80:
            result= "Your academic standing is 'Highest Distinction'"
        elif self.cgpa>3.65:
            result= "Your academic standing is 'High Distinction'"
        elif self.cgpa>3.50:
            result= "Your academic standing is 'Distinction'"
        elif self.cgpa> 2.00:
            result= "Your academic standing is 'Satisfactory'"
        else:
            result= "Sorry, you Cannot graduate"    

        print(f"Name: {self.name}, ID: {self.id}\nDepartment: {self.dept}\nCGPA: {self.cgpa:.2f}\n{result}")            
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
#TASK 8
class Shinobi:
    def __init__(self, name, rank):
        self.salary= 0
        self.mission= 0
        self.name= name
        self.rank= rank
        
    def changeRank(self, rank):
        self.rank= rank

    def calSalary(self, mission):
        self.mission= mission
        
        if self.rank== "Genin":
            self.salary= self.mission*50
        elif self.rank== "Chunin":
            self.salary= self.mission*100
        else:
            self.salary= self.mission*500

    def printInfo(self):
        print(f"Name: {self.name}\nRank: {self.rank}\nNumber of mission: {self.mission}\
        \nSalary: {self.salary}")
    
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()    
```

```python
#TASK 10
class UberEats:
    def __init__(self, name, ph_no, place):
        self.name= name
        self.ph_no= ph_no
        self.place= place
        print(f"{self.name} welcome to UberEats!")

    def add_items(self, item_1, item_2, price_1, price_2):
        self.price_1= price_1
        self.price_2= price_2
        self.dict_order= {}
        self.dict_order[item_1]= self.price_1
        self.dict_order[item_2]= self.price_2 


    def print_order_detail(self):
        return f"User details: Name: {self.name}, Phone:\n{self.ph_no}, Address: {self.place}\nOrders: {self.dict_order}\nTotal Paid Amount: {self.price_1+self.price_2}"     


order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```

```python
#TASK 11
class Spotify:
    def __init__(self, list_1):
        self.list_1= list_1
        self.len= len(list_1)
        print("Welcome to Spotify!")

    def add_to_playlist(self, name):
        self.list_1.append(name)   
        self.len+= 1

    def playing_number(self, num):

        if num>self.len:
            return f"{num} number song not found. Your playlist\nhas {self.len} songs only."
        else:
            return f"##########################\nPlaying {num} number song for you\nSong name: {self.list_1[num-1]}"
        
            

user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
